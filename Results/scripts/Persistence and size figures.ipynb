```python
lengths = []

with open(dropbox + "persistence.csv") as f:
    next(f)
    for l in f:
        toks = l.strip("\n").split(",")
        
        lengths.append(int(toks[1]))
```

```python
sizes = []

with open(dropbox + "epidemic_size.csv") as f:
    next(f)
    for l in f:
        toks = l.strip("\n").split(",")
        
        sizes.append(int(toks[1]))
```

```python
size_count = Counter(sizes)
length_count = Counter(lengths)
```

```python
grouped_sizes = {}

grouped_sizes["<100"] = 0
grouped_sizes["100-500"] = 0
grouped_sizes["500-1000"] = 0
grouped_sizes["1000-5000"] = 0
grouped_sizes["1.5m-1.6m"] = 0
grouped_sizes["1.6m-1.7m"] = 0
grouped_sizes["1.7m-1.8m"] = 0
grouped_sizes["1.8m-1.9m"] = 0
grouped_sizes[">1.9m"] = 0

for k,v in size_count.items():
    if k > 0 and k < 100:
        grouped_sizes["<100"] += v
    elif k > 100 and k < 500:
        grouped_sizes["100-500"] += v
    elif k > 500 and k < 1000:
        grouped_sizes["500-1000"] += v
    elif k > 1000 and k < 5000:
        grouped_sizes["1000-5000"] += v
    elif k > 1500000 and k < 1600000:
        grouped_sizes["1.5m-1.6m"] += v
    elif k > 1600000 and k < 1700000:
        grouped_sizes["1.6m-1.7m"] += v
    elif k > 1700000 and k < 1800000:
        grouped_sizes["1.7m-1.8m"] += v
    elif k > 1800000 and k < 1900000:
        grouped_sizes["1.8m-1.9m"] += v
    elif k > 19000000:
        grouped_sizes[">1.9m"] += v
        
    else:
        print(k)

```

```python
fig,ax = plt.subplots(1,1, figsize=(15,15))

x = list(grouped_sizes.keys())
y = list(grouped_sizes.values())

print(x,y)

plt.bar(x,y)

plt.xticks(size=25, rotation=90)
plt.yticks(size=20)

plt.xlabel("Case count", size=25)
plt.ylabel("Frequency", size=25)

plt.tight_layout()
plt.savefig(dropbox + "size_figure.pdf", format='pdf')



plt.show()
```
