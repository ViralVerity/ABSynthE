```python
def plot_R0(input_file, title):

    R0_list = []
    R0_freqs = Counter()


    with open(input_file) as f:
        for l in f:
            R0 = l.strip("\n")
            R0_list.append(round(float(R0), 2))


    for i in R0_list:
        R0_freqs[i] += 1
        
    #print(R0_freqs)

#     x = R0_freqs.keys()
#     y = R0_freqs.values()

    x = []
    y = []

    for key, value in R0_freqs.items():
        x.append(key)
        y.append(value)
    
    width = 1/75

    plt.title(title)
    plt.xlabel("R0")
    plt.ylabel("Frequency")
    #plt.xlim(min(x), max(x))
    
    plt.bar(x, y, width, alpha = 0.5)
    plt.show()

#     n, bins, patches = plt.hist(x = x, bins = 'auto',
#                                color='#0504aa',
#                                 alpha=0.7, rwidth=0.85)
    plt.show()
    
    

    # print(max(x))

    # print(x)
```

```python
def plot_size(input_file, title):
    
    size_list = []
    size_freqs = Counter()


    with open(input_file) as f:
        for l in f:
            size = l.strip("\n").split(",")[0]
            size_list.append(int(size))

    for i in size_list:
        size_freqs[i] += 1

    x2 = []
    y2 = []

    for key, value in size_freqs.items():
        x2.append(key)
        y2.append(value)
        
    plt.bar(x2,y2)


#     n, bins, patches = plt.hist(x = x2, bins = 'auto',
#                                color='#0504aa',
#                                 alpha=0.7, rwidth=0.85)
    
#     plt.show()


    plt.title(title)
    plt.xlabel("Size of epidemic")
    plt.ylabel("frequency")
    plt.show()
```

```python
def plot_persistence(input_file):
    
    len_list = []
    len_freqs = []
    
    with open(input_file) as f:
        for l in f:
            length = l.strip("\n").split(",")[0]
            len_list.append(int(length))
            

    for i in len_list:
        len_freqs[i] += 1

    x = []
    y = []

    for key, value in len_freqs.items():
        x.append(key)
        y.append(value)
        
    plt.bar(x,y)


#     n, bins, patches = plt.hist(x = x2, bins = 'auto',
#                                color='#0504aa',
#                                 alpha=0.7, rwidth=0.85)
    
#     plt.show()


    plt.title(title)
    plt.xlabel("Length of Epidemic (days)")
    plt.ylabel("Frequency")
    plt.show()

```

```python
def plot_spread(direc_big):

    district_list = ["bo", 'bombali', 'bonthe', 'kailahun', 'kambia', 'kenema', 'koinadugu', 'kono', 'moyamba', 'portloko', 'pujehun', 'tonkolili', 'westernarearural', 'westernareaurban']

    dist_count = {}

    for dist in district_list:
        dist_count[dist] = 0

    file_count = 0

    for directory1 in os.listdir(direc_big):
        print(directory1)
        if directory1 != ".DS_Store":
            for log_file in os.listdir(direc_big + directory1):
                if log_file != "large_epidemics.csv" and log_file != ".DS_Store":
                    #print(log_file)
                    dist_set = set()
                    file_count += 1
                    with open(direc_big + directory1 + "/" + log_file) as f:
                        next(f)
                        for l in f:
                            toks = l.strip("\n").split(",")
                            #print(toks)
                            district = toks[3]

                            dist_set.add(district)

                    for dist in district_list:
                        if dist in dist_set:
                            #print(dist)
                            #print(dist_count[dist])
                            dist_count[dist] += 1  
                            #print(dist_count[dist])

                    #print(dist_count)

    dist_percentage = {}

    for key, count in dist_count.items():
        dist_percentage[key] = (count/file_count)*100


    x = [i for i in range(14)]
    my_xticks = district_list
    
    plt.xticks(x,my_xticks, rotation=90)
    
    y = list(dist_percentage.values())

    plt.title("Percentage of times district was involved in epidemic")
    
    plt.bar(x,y)
    
    plt.savefig(path + results + "from_server/spread_across_districts.pdf", bbox_inches="tight")
    
    plt.show()
```

```python
plot_spread(path + results + "from_server/all_logs/")
```

```python
def plot_all_intermediate(directory):
    
    plt.rcParams.update({'font.size': 25})
        
    plot_spread(directory)
    
    size_list = []
    size_freqs = Counter()

    for input_file in os.listdir(directory):
        with open(directory + input_file) as f:
            next(f)
            size = 0
            for l in f:
                size += 1
            
            size_list.append(size)

    for i in size_list:
        size_freqs[i] += 1

    x2 = list(size_freqs.keys())
    y2 = list(size_freqs.values())

#     print(size_freqs)
    
#     for key, value in size_freqs.items():
#         x2.append(key)
#         y2.append(value)
    
#     print(x2, y2)
    
    plt.scatter(x2,y2)

    plt.xticks(rotation=45)
    
    plt.title("Epidemic sizes")
    plt.xlabel("Size of epidemic")
    plt.ylabel("Frequency")
    
    plt.savefig(path+results+"2/epidemics_sizes.pdf", bbox_inches="tight")
    
    plt.show()
    
    
    
    
    len_list = []
    len_freqs = Counter()
    
    for input_file in os.listdir(directory):
        day_list = []
        with open(directory + input_file) as f:
            next(f)
            for l in f:
                day = l.strip("\n").split(",")[5]
                day_list.append(int(day))
                
            length = max(day_list)
            len_list.append(length)
            
    for i in len_list:
        len_freqs[i] += 1

    x3 = []
    y3 = []

    for key, value in len_freqs.items():
        x3.append(key)
        y3.append(value)
        
    #print(x,y)
        
    #plt.bar(x3,y3)


    n, bins, patches = plt.hist(x = x3, bins = 'auto',
                               color='#0504aa',
                                alpha=0.7, rwidth=0.85)
    
#     plt.show()


    plt.title("Length of Epidemics")
    plt.xlabel("Length of Epidemic (days)")
    plt.ylabel("Frequency")
    
    plt.savefig(path+results+"2/persistence.pdf", bbox_inches="tight")
    
    plt.show()




    
```
